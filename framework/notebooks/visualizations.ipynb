```python
import scanpy as sc
adata = sc.read("../../data/preprocessed/adamson_preprocessed_0.h5ad")
```

```python
sc.tl.pca(adata, svd_solver='arpack')
sc.pp.neighbors(adata, n_neighbors=15, n_pcs=40)
sc.tl.umap(adata)
```

```python
sc.pl.umap(adata, color="perturbation")
```

```python
# looking at other possible sources of variation (other than perturbation)
sc.pl.umap(adata, color=['ncounts', 'nperts'], cmap='viridis')
```
